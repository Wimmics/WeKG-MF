## **Preprocessing step**

Météo France provides open access to data observations. You can download different SYNOP observations files from https://donneespubliques.meteofrance.fr/?fond=produit&id_produit=90&id_rubrique=32. 

We developped this jupyter script to perform a simple preprocessing required to have the good format that should enable the upload of data later in mongoDB database. 

```python
import pandas as pd
from pandas import DataFrame
```

```python
values = list()
file_obs = open("synop202108.csv", "r")
for line in file_obs.readlines(): 
  values.append(line.split(";"))

column_label = values[0]
print(column_label)

```

```python
def synop_to_df(filepath): 
  values = list()
  file_obs = open(filepath, "r")
  for line in file_obs.readlines(): 
    values.append(line.split(";"))
  column_label = values[0]
  del values[0]
  df = DataFrame(values, columns=column_label)
  df = df.drop(['\n'], axis=1)
  return df

```

```python
df = synop_to_df("synop202108.csv")
print(df.shape)
```

```python
df["dateF"] = pd.to_datetime(df["date"], infer_datetime_format=True)
```
